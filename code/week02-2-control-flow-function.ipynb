```python
value = "apple"

match value:
  case 'apple':
    result = "사과"
  case 'banana':
    result = "바나나"
  
print(result)
```

```python
value = "banana"

match value:
  case 'apple':
    result = "사과"
  case 'banana':
    result = "바나나"
  
print(result)
```

```python
value = "mango"

match value:
  case 'apple':
    result = "사과"
  case 'banana':
    result = "바나나"
  case _:
    result = None
  
print(result)
```

```python
def sum_sub(x, y):
  hap = x + y
  cha = x - y
  return hap, cha

sum_sub(10, 3)
```

```python
sum_sub(x = 10, y = 3)
```

```python
sum_sub(y = 3, x = 10)
```
