# 서브플롯 간의 간격 자동 조정
- constrained_layout=True로 설정
    - 서브플롯들 사이의 간격을 자동으로 조정합니다. 이로 인해 서브플롯의 제목, 축 라벨, 눈금 등이 서로 겹치지 않게 됩니다.

```python
import matplotlib.pyplot as plt

# 2x2 서브플롯 생성
fig, axs = plt.subplots(ncols=2, nrows=2, constrained_layout=True)

# 서브플롯에 그래프 추가
for i in range(2):
    for j in range(2):
        axs[i, j].plot([1, 2, 3], [1, 4, 9])

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

fig4 = plt.figure(constrained_layout=True)
spec4 = fig4.add_gridspec(ncols=2, nrows=2)

anno_opts = dict(xy=(0.5, 0.5), xycoords='axes fraction',
                 va='center', ha='center')

f4_ax1 = fig4.add_subplot(spec4[0, 0])
f4_ax1.annotate('GridSpec[0, 0]', **anno_opts)
fig4.add_subplot(spec4[0, 1]).annotate('GridSpec[0, 1:]', **anno_opts)
fig4.add_subplot(spec4[1, 0]).annotate('GridSpec[1:, 0]', **anno_opts)
fig4.add_subplot(spec4[1, 1]).annotate('GridSpec[1:, 1:]', **anno_opts);
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

fig5 = plt.figure(constrained_layout=True)
widths = [2, 3, 1.5]
heights = [1, 3, 2]

spec5 = fig5.add_gridspec(ncols=3, nrows=3, 
                          width_ratios=widths, height_ratios=heights)
for row in range(3):
    for col in range(3):
        ax = fig5.add_subplot(spec5[row, col])
        label = 'Width: {}\nHeight: {}'.format(widths[col], heights[row])
        ax.annotate(label, (0.1, 0.5), xycoords='axes fraction', va='center')
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

widths = [4, 6, 3]
heights = [1, 3, 2]
gs_kw = dict(width_ratios=widths, height_ratios=heights)

fig6, f6_axes = plt.subplots(ncols=3, nrows=3, 
                             constrained_layout=True, gridspec_kw=gs_kw)

for r, row in enumerate(f6_axes):
    for c, ax in enumerate(row):
        label = 'Width: {}\nHeight: {}'.format(widths[c], heights[r])
        ax.annotate(label, (0.1, 0.5), xycoords='axes fraction', va='center')
```

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

def squiggle_xy(a, b, c, d, i=np.arange(0.0, 2*np.pi, 0.05)):
    return np.sin(i*a)*np.cos(i*b), np.sin(i*c)*np.cos(i*d)

myfig = plt.figure(constrained_layout=True)
mygs = myfig.add_gridspec(1, 2) # 1행 2열

subgs0 = mygs[0].subgridspec(2, 3) # 1열 내부에 2행 3열의 서브플롯 
subgs1 = mygs[1].subgridspec(3, 2) # 2열 내부에 3행 2열의 서브플롯 

ax0 = subgs0.subplots() # Create all subplots for the inner grid.
a, b = 0, 0
for(c, d), ax in np.ndenumerate(ax0):
    ax.set_title(f'{a} {b}, {c}, {d}')
    ax.plot(*squiggle_xy(a + 1, b + 1, c + 1, d + 1))
    ax.set(xticks=[], yticks=[])

ax1 = subgs1.subplots() # Create all subplots for the inner grid.
a, b = 0, 1
for(c, d), ax in np.ndenumerate(ax1):
    ax.set_title(f'a={a} b={b}, c={c}, d={d}')
    ax.plot(*squiggle_xy(a + 1, b + 1, c + 1, d + 1))
    ax.set(xticks=[], yticks=[])

plt.show()
```

```python
# chatgpt code

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# GridSpec으로 3행 3열의 그리드 생성
fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(3, 3)

# 첫 번째 서브플롯 생성
ax1 = fig.add_subplot(gs[0, 0])
ax1.set_title('Subplot 1')

# 두 번째 서브플롯 생성
ax2 = fig.add_subplot(gs[0, 1])
ax2.set_title('Subplot 2')

# 세 번째 서브플롯 생성
ax3 = fig.add_subplot(gs[0, 2])
ax3.set_title('Subplot 3')

# 네 번째 서브플롯 생성
ax4 = fig.add_subplot(gs[1, :])
ax4.set_title('Subplot 4: Merged across second row')

# `get_gridspec()`을 사용해 네 번째 서브플롯의 GridSpec 반환 및 수정
ax4.get_gridspec().update(wspace=0.3, hspace=0.9)

gd3 = ax3.get_gridspec()
gd4 = ax4.get_gridspec()

print(id(gd3), id(gd4))
print(id(ax1), id(ax2))

# plt.tight_layout()
plt.show()
```
